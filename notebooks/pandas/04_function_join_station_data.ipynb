### 🚀 Step 1: Import Required Libraries

First, let's import the library we need:

````python
import pandas as pd

print(f"✅ Pandas version: {pd.__version__}")
print("📚 Ready to join datasets!")
````

### 📊 Step 2: Load Both Datasets

Let's load the station metadata and the temperature readings datasets:

````python
stations_file = '../../data/weather_stations.csv'
readings_file = '../../data/temperature_readings.csv'

stations_df = pd.read_csv(stations_file)
readings_df = pd.read_csv(readings_file)

print("✅ Files loaded successfully!")
print(f"📁 Stations file: {stations_file}")
print(f"📁 Readings file: {readings_file}")
````

### 🔍 Step 3: Explore the Two DataFrames

Before joining, let's understand what each table contains:

````python
print("📍 STATIONS DATAFRAME")
print(f"Shape: {stations_df.shape}")
print(f"Columns: {list(stations_df.columns)}")
display(stations_df.head())

print("\n" + "="*70 + "\n")

print("🌡️ READINGS DATAFRAME")
print(f"Shape: {readings_df.shape}")
print(f"Columns: {list(readings_df.columns)}")
display(readings_df.head())
````

### 🔑 Step 4: Understand the Join Key

A join works by matching rows from both tables using a common field. In this case, the shared key is `station_id`:

````python
print("📍 Station IDs in stations_df:")
print(sorted(stations_df['station_id'].unique()))

print("\n🌡️ Station IDs in readings_df:")
print(sorted(readings_df['station_id'].unique()))

stations_ids = set(stations_df['station_id'].unique())
readings_ids = set(readings_df['station_id'].unique())

print("\n✅ Station IDs in both datasets:")
print(sorted(stations_ids.intersection(readings_ids)))
````

### 🔗 Step 5: Perform a Basic Merge

Now let's join the two DataFrames using `pd.merge()`:

````python
joined_df = pd.merge(readings_df, stations_df, on='station_id', how='left')

print("✅ Merge complete!")
print(f"📏 Joined shape: {joined_df.shape}")
print(f"📋 Joined columns: {list(joined_df.columns)}")
display(joined_df.head())
````

### 🧪 Step 6: Understand Left Join Behavior

A left join keeps all rows from the readings table and adds matching station information where available:

````python
print(f"Original readings rows: {len(readings_df)}")
print(f"Joined rows: {len(joined_df)}")

if len(joined_df) == len(readings_df):
    print("✅ All reading rows were preserved in the join")
else:
    print("⚠️ Row count changed during join")

print("\n🔍 Missing joined values check:")
missing_counts = joined_df.isnull().sum()
print(missing_counts[missing_counts > 0])
````

````python
def join_station_data(stations_df, readings_df):
    """
    Join station metadata with environmental readings.
    
    Args:
        stations_df (pandas.DataFrame): Station metadata
        readings_df (pandas.DataFrame): Environmental readings
        
    Returns:
        pandas.DataFrame: Joined dataset
    """
    
    print("=" * 50)
    print("JOINING STATION DATA")
    print("=" * 50)
    
    # Check for empty or missing input
    if stations_df is None or stations_df.empty:
        print("❌ ERROR: Empty or None stations DataFrame provided")
        return pd.DataFrame()
    
    if readings_df is None or readings_df.empty:
        print("❌ ERROR: Empty or None readings DataFrame provided")
        return pd.DataFrame()
    
    # Check for required join column
    if 'station_id' not in stations_df.columns:
        print("❌ ERROR: 'station_id' column missing from stations DataFrame")
        return pd.DataFrame()
    
    if 'station_id' not in readings_df.columns:
        print("❌ ERROR: 'station_id' column missing from readings DataFrame")
        return pd.DataFrame()
    
    print(f"📍 Stations rows: {len(stations_df)}")
    print(f"🌡️ Readings rows: {len(readings_df)}")
    
    # Perform left join
    joined_df = pd.merge(readings_df, stations_df, on='station_id', how='left')
    
    print(f"\n📏 Joined rows: {len(joined_df)}")
    print(f"📋 Joined columns: {list(joined_df.columns)}")
    
    print("\n✅ Station data joined successfully!")
    
    return joined_df
````
