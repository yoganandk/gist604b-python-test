### 📚 Step 1: Understanding Vector Data Formats

GeoPandas can read many spatial formats:

| Format | Extension | Pros | Cons |
|--------|-----------|------|------|
| **GeoJSON** | `.geojson` | Web-friendly, human-readable, single file | Larger file size |
| **Shapefile** | `.shp` | Industry standard | Multiple files required |
| **GeoPackage** | `.gpkg` | Modern, efficient, single file | Newer format |

**This assignment uses GeoJSON** for all datasets.

### 💻 Step 2: Basic Loading - The Simple Way

Let's start by loading data the simple way:

````python
import geopandas as gpd

# Load cities data directly
cities = gpd.read_file('../../data/cities/ne_cities_us.geojson')

print(f"✅ Loaded {len(cities)} city features")
print(f"📍 CRS: {cities.crs}")
print(f"📊 Columns: {list(cities.columns)[:5]}...")  # Show first 5 columns
cities.head(3)
````

### 🗺️ Step 3: Visualization 1 - Show Me the Data

Now that we've loaded data, let's **visualize** it! This is where GIS gets exciting - we can see our data on a map.

**Why visualize?**
- ✅ Verify data loaded correctly
- ✅ Understand spatial distribution
- ✅ Catch obvious errors (wrong location, missing features)

GeoPandas makes plotting spatial data incredibly easy with `.plot()`!

- **Points** represent discrete locations (cities, observation sites)
- **Polygons** represent areas (regions, parks, zones)
- **Both** are loaded the same way with `gpd.read_file()`
- `gpd.read_file()` returns a `GeoDataFrame` - this is GeoPandas' main data structure!
- **Visualization helps** you understand your data structure immediately

Let's visualize **points** vs **polygons**:


````python
# Load ecoregions for comparison
ecoregions = gpd.read_file('../../data/ecoregions/epa_level3_western_us.geojson')

# Create side-by-side comparison
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))

# LEFT: Point data (cities)
cities.plot(ax=ax1, color='red', markersize=30, alpha=0.7, edgecolor='darkred', linewidth=0.5)
ax1.set_title(f"Point Data: {len(cities)} Cities\nGeometry Type: {cities.geometry.geom_type.iloc[0]}", 
             fontsize=12, fontweight='bold')
ax1.set_xlabel("Longitude")
ax1.set_ylabel("Latitude")
ax1.grid(True, alpha=0.3)

# RIGHT: Polygon data (ecoregions)
ecoregions.plot(ax=ax2, alpha=0.6, edgecolor='black', linewidth=1, cmap='Set3')
ax2.set_title(f"Polygon Data: {len(ecoregions)} Ecoregions\nGeometry Type: {ecoregions.geometry.geom_type.iloc[0]}", 
             fontsize=12, fontweight='bold')
ax2.set_xlabel("Longitude")
ax2.set_ylabel("Latitude")
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("📊 Comparison:")
print(f"   Points (cities): {len(cities)} discrete locations")
print(f"   Polygons (ecoregions): {len(ecoregions)} area features")
print("   Both can be loaded the same way with gpd.read_file()!")
````

### 🛠️ Step 5: Working with Path Objects

Professional code uses `pathlib.Path` for cross-platform file path handling:

**Why Path objects?**
- Work on Windows, Mac, Linux
- Easy file existence checking with `.exists()`
- Clear path manipulation
- Built-in validation methods

````python
from pathlib import Path

# String path (what users might provide)
string_path = '../../data/cities/ne_cities_us.geojson'

# 💡 Convert strings to Path objects - works with both strings and Path inputs
path_obj = Path(string_path)

print(f"String path: {string_path}")
print(f"Path object: {path_obj}")
print(f"Absolute path: {path_obj.resolve()}")
print(f"File exists? {path_obj.exists()}")
print(f"File name: {path_obj.name}")
print(f"File suffix: {path_obj.suffix}")
````

### ⚠️ Step 6: Understanding Error Handling - File Not Found

What happens when a file doesn't exist?

**💡 This pattern will be used in our function!**

````python
# 💡 Check if file exists BEFORE trying to load
file_path = Path('../../data/cities/ne_cities_us.geojson')

if not file_path.exists():
    # 💡 Raise FileNotFoundError with helpful message
    print(f"Would raise: FileNotFoundError('File not found: {file_path}')")
else:
    print("✅ File exists, safe to load")
    data = gpd.read_file(file_path)
    print(f"✅ Successfully loaded {len(data)} features")
````

### 🔥 Step 7: Using try/except for Robust Error Handling

Even if a file exists, it might not be readable (corrupted, wrong format, etc.).

**💡 This try/except pattern will be used in our function!**

````python
# 💡 Professional error handling with try/except
file_path = Path('../../data/cities/ne_cities_us.geojson')

try:
    # Try to load the file
    gdf = gpd.read_file(file_path)
    print(f"✅ Success! Loaded {len(gdf)} features")
    print(f"📍 CRS: {gdf.crs}")
    
except Exception as e:
    # If anything goes wrong, catch the error
    # 💡 Convert to ValueError with helpful message including original error
    print(f"Would raise: ValueError('Cannot read spatial data from {file_path}: {str(e)}')")
````

### 🎯 Step 8: Understanding **kwargs (Keyword Arguments)

**What is `**kwargs`?**
- Stands for "keyword arguments"
- Allows passing any number of named parameters
- Forward them with `**kwargs`

```python
def my_function(required_arg, **kwargs):
    # Forward all extra arguments to another function
    result = gpd.read_file(required_arg, **kwargs)
    return result
```
Our function signature includes `**kwargs` - this allows passing extra arguments to `gpd.read_file()`:

**💡 This will make our function flexible!**

````python
# 💡 gpd.read_file() accepts optional parameters - we'll pass them through

# Example: Load only first 5 rows (rows is an optional parameter)
cities_small = gpd.read_file('../../data/cities/ne_cities_us.geojson', rows=5)
print(f"Loaded only {len(cities_small)} cities (using rows=5)")

# Example: Load specific columns (columns is an optional parameter)
cities_minimal = gpd.read_file(
    '../../data/cities/ne_cities_us.geojson', 
    columns=['NAME', 'geometry']
)
print(f"Loaded {len(cities_minimal.columns)} columns: {list(cities_minimal.columns)}")
````

### 🏗️ Step 9: Building the Complete Function

Now let's put everything together into a reusable function. This is what you will implement in `src/geopandas_basics.py`.

**Notice how it uses:**
- ✅ `Path()` to handle both strings and Path objects
- ✅ `.exists()` check with `FileNotFoundError`
- ✅ `try/except` block for error handling
- ✅ `**kwargs` to forward optional parameters

````python
from typing import Union
import geopandas as gpd
from pathlib import Path

def load_spatial_data(file_path: Union[str, Path], **kwargs) -> gpd.GeoDataFrame:
    """
    Load spatial data from various vector file formats.
    
    Supports GeoJSON, Shapefile, GeoPackage, and other formats readable by GeoPandas.
    Professional implementation with comprehensive error handling.
    
    Args:
        file_path: Path to the spatial data file (string or Path object)
        **kwargs: Additional arguments to pass to gpd.read_file()
        
    Returns:
        GeoDataFrame containing the loaded spatial data
        
    Raises:
        FileNotFoundError: If the file doesn't exist
        ValueError: If the file format is invalid or cannot be read
        
    Example:
        >>> gdf = load_spatial_data('data/cities.geojson')
        >>> print(f"Loaded {len(gdf)} features")
    """
    # Convert to Path object for consistent handling
    file_path = Path(file_path)
    
    # Check if file exists
    if not file_path.exists():
        raise FileNotFoundError(f"File not found: {file_path}")
    
    # Try to load the spatial data
    try:
        gdf = gpd.read_file(file_path, **kwargs)
    except Exception as e:
        raise ValueError(f"Cannot read spatial data from {file_path}: {str(e)}")
    
    return gdf
````
