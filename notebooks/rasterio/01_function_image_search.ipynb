### 📚 Step 1: Import Libraries

Let's import the libraries needed to search for satellite imagery:

````python
from pystac_client import Client

# STAC API endpoint
stac_url = "https://planetarycomputer.microsoft.com/api/stac/v1"

print(f"✅ STAC URL set: {stac_url}")
````

### 🌍 Step 2: Define Search Parameters

We need a bounding box, a date range, and a cloud cover threshold before searching.

````python
# Bounding box: [west, south, east, north]
bbox = [-110.74772571639987, 32.270431012618026, -110.70996215904789, 32.29386169894274]

# Date range
date_range = "2021-07-05/2021-08-02"

# Cloud cover threshold
scene_cloud_tolerance = 5

print(f"Bounding box: {bbox}")
print(f"Date range: {date_range}")
print(f"Cloud tolerance: {scene_cloud_tolerance}%")
````

### 🔎 Step 3: Connect to the STAC Catalog

**💡 This will be used in our function!**

````python
catalog = Client.open(stac_url)
print(f"✅ Connected to catalog: {type(catalog)}")
````

### 🛰️ Step 4: Search the Catalog

**💡 This will be used in our function!**

````python
search = catalog.search(
    collections=["sentinel-2-l2a"],
    bbox=bbox,
    datetime=date_range,
    query={"eo:cloud_cover": {"lt": scene_cloud_tolerance}}
)

items = list(search.items())
print(f"✅ Found {len(items)} items")
````

### 🗂️ Step 5: Explore Search Results

Let's inspect the STAC items returned by our search.

````python
item = items[0]

print(f"Item ID: {item.id}")
print(f"Datetime: {item.datetime}")
print(f"Bounding box: {item.bbox}")
print(f"Cloud cover: {item.properties.get('eo:cloud_cover', 'N/A')}%")
````

### 🔗 Step 6: Inspect Scene Assets

Each STAC item contains assets such as red, nir, and visual imagery. These assets store the URLs we will use later.

````python
print("Available assets:")
print(list(item.assets.keys()))
print()

print("Visual asset:")
print(item.assets["visual"])
print()

print("Red band URL:")
print(item.assets["B04"].href)
print()

print("NIR band URL:")
print(item.assets["B08"].href)
````

### 🧩 Step 7: Building the Complete Function

Now let's put everything together into a reusable function. This is what you will implement in `src/rasterio_basics.py`.

````python
from pystac_client import Client

stac_url = "https://planetarycomputer.microsoft.com/api/stac/v1"

def image_search(bbox, date_range, scene_cloud_tolerance):
    """
    Search for Sentinel-2 scenes using a STAC catalog.
    
    Args:
        bbox: [west, south, east, north]
        date_range: Date range in "YYYY-MM-DD/YYYY-MM-DD" format
        scene_cloud_tolerance: Maximum allowed cloud cover percentage
        
    Returns:
        List of matching STAC items
    """
    catalog = Client.open(stac_url)
    
    search = catalog.search(
        collections=["sentinel-2-l2a"],
        bbox=bbox,
        datetime=date_range,
        query={"eo:cloud_cover": {"lt": scene_cloud_tolerance}}
    )
    
    return list(search.items())
````
